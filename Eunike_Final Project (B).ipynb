```python
# a
data.dropna(thresh=0.5*len(data), axis=1, inplace=False)
```

```python
# b
data.loc[:, "agent"].fillna(data.loc[:, "agent"].mean())
```

```python
# c
data.loc[:, "country"].fillna(data.loc[:, "country"].mode()[0])
```

```python
data.loc[:, "children"].fillna(data.loc[:, "children"].mode()[0])
```

```python
data["category"] = pd.cut(data["lead_time"], bins=3, labels=["Short", "Medium", "Long"])
data.loc[:, ["lead_time", "category"]]
```

```python
pd.to_datetime(data["reservation_status_date"], format='%Y-%m-%d')
data["reservation_status_date"]
```

```python
filtered_data = data.loc[(data["reservation_status_date"] > '2017-12-31')]
filtered_data
```

```python
# Konversi kolom menjadi list
arrival_date_month_list = data["arrival_date_month"].tolist()
print(arrival_date_month_list)
```

```python
# Plotting
plt.figure(figsize=(12,12))
plt.hist(arrival_date_month_list, color="r", rwidth=0.8)
plt.title("Arrival Date Month", fontdict = {"fontsize": 20})

plt.xlabel("Month", fontdict = {"fontsize": 18})
plt.ylabel("Number of Visitors", fontdict = {"fontsize": 18})

plt.grid(axis="y", ls="--", alpha=0.8)

plt.show()
```
